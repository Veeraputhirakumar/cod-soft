```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_data = pd.read_csv('IMDb Movies India.csv')
```

```python
movie_data
```

```python
# Scatter Plot of Movie Duration vs. Rating
sns.scatterplot(data=movie_data, x="Duration", y="Rating")
plt.xlabel("Movie Duration (minutes)")
plt.ylabel("Movie Rating")
plt.title("Movie Duration vs. Rating")
plt.show()
```

```python
# Bar Chart: Average Rating by Genre
average_ratings_by_genre = movie_data.groupby("Genre")["Rating"].mean().reset_index()
sns.barplot(x="Genre", y="Rating", data=average_ratings_by_genre)
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.title(" Average Rating by Genre")
plt.xticks(rotation=90)
plt.show()
```

```python
# Pair Plot for Numeric Features
numeric_features = ["Duration", "Votes", "Rating"]
sns.pairplot(movie_data[numeric_features])
plt.show()
```

```python
# Box Plot: Rating by Director
plt.figure(figsize=(10, 6))
sns.boxplot(x="Director", y="Rating", data=movie_data)
plt.xlabel("Director")
plt.ylabel("Movie Rating")
plt.title("Rating by director")
plt.xticks(rotation=90)
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = movie_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

```python
# Box Plot: Rating by Director
plt.figure(figsize=(12, 6))
sns.boxplot(x="Director", y="Rating", data=movie_data)
plt.xlabel("Director")
plt.ylabel("Movie Rating")
plt.title("Box Plot: Rating by Director")
plt.xticks(rotation=90)
plt.show()
```

```python
# Count Plot: Number of Movies by Year
plt.figure(figsize=(12, 6))
sns.countplot(x="Year", data=movie_data)
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.title("Number of Movies by Year")
plt.xticks(rotation=45)
plt.show()
```

```python
# Violin Plot: Rating Distribution by Genre
plt.figure(figsize=(12, 6))
sns.violinplot(x="Genre", y="Rating", data=movie_data)
plt.xlabel("Genre")
plt.ylabel("Movie Rating")
plt.title("Rating Distribution by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count Plot: Number of Movies by Genre
plt.figure(figsize=(12, 6))
sns.countplot(x="Genre", data=movie_data)
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.title("Number of Movies by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot: Votes vs. Rating
sns.scatterplot(data=movie_data, x="Votes", y="Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Movie Rating")
plt.title("Scatter Plot: Votes vs. Rating")
plt.show()
```
